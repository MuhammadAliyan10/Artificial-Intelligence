```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Data/daily_water_usage.csv')
df.head()
```

```python
X = df[['temperature',"number_of_fasters","is_weekend"]]
y = df['water_usage']
```

```python
model =LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df["temperature"],
    df["water_usage"],
    c=df["is_weekend"],
    s=df["number_of_fasters"] * 50,
    alpha=0.6,
    cmap="viridis",

    label="Real Data"
)

plt.scatter(
    df["temperature"],
    model.predict(X),
    color="red",
    s=20,
    label="Predicted Water Usage",
    alpha=0.5
)


plt.xlabel("Temperature (°C)")
plt.ylabel("Water Usage (liters)")
plt.title("Water Usage vs. Temperature with Fasters and Weekend Effects")
plt.colorbar(scatter, label="Is Weekend (0 = No, 1 = Yes)")
plt.legend()


plt.show()
```
