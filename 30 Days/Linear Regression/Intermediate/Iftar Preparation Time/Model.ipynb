```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Data/iftar_preparation_time.csv')
df.head()
```

```python
X = df[['number_of_dishes',"number_of_helpers"]]
y = df['preparation_time']
```

```python
model =LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
mask = df["number_of_helpers"] == 1
X_filtered = X[mask]
y_filtered = y[mask]
dishes_filtered = X_filtered["number_of_dishes"]

plt.scatter(dishes_filtered, y_filtered, color="red", label="Real Data (Helpers = 1)")
plt.plot(dishes_filtered, model.predict(X_filtered), color="blue", label="Regression Line")
plt.xlabel("Number of Dishes")
plt.ylabel("Preparation Time (minutes)")
plt.title("Iftar Prep Time vs. Number of Dishes (Fixed Helpers = 1)")
plt.legend()
plt.show()
```
