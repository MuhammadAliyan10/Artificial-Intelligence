```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./Data/prayer_time.csv')
df.head()
```

```python
X = df[['days']]
y = df['prayer_time']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
slope = model.coef_
slope
```

```python
intercept = model.intercept_
intercept
```

```python
plt.scatter(X, y, color="blue",label="Real Data")
plt.plot(X, model.predict(X), color="red", label="Regression")
plt.xlabel('Day of Ramadan')
plt.ylabel('Prayer Time (minutes)')
plt.legend()
plt.show()
```
