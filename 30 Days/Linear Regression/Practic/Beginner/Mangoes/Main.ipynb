```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv('./Data/mango_prices.csv')
df.head()
```

```python
X = df[['amount_mangoes']]
y = df['price']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X, y, color='blue', label='Real Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Number of Mangoes')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Mango Prices')
plt.legend()
plt.show()
```
