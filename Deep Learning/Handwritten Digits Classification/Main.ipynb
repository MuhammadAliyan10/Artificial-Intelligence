```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
plt.matshow(x_test[0])
```

```python
y_test[0]
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_test[0]
```

```python
x_train_fa = x_train.reshape(len(x_train),28*28)
x_test_fa = x_test.reshape(len(x_test),28*28)
x_train_fa.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_fa,y_train,epochs=5)
```

```python
model.evaluate(x_test_fa,y_test)
```

```python
plt.matshow(x_test[1])
```

```python
y_predicted = model.predict(x_test_fa)
```

```python
y_predicted[1]
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
```

```python
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train_fa,y_train,epochs=5)
```

```python
model.evaluate(x_test_fa,y_test)
```

```python
y_predicted = model.predict(x_test_fa)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
```

```python
plt.figure(figsize=(10,6))
snn.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=5)
```

```python
y_predicted = model.predict(x_test)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
plt.figure(figsize=(10,6))
snn.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
