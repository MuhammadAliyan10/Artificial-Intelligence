```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv('./Data/titanic.csv')
df.head(30)
```

```python
df = df.dropna()
```

```python
df.head(30)
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df
```

```python
sex_l = LabelEncoder()
df['sex_n'] =sex_l.fit_transform(df['Sex'])
```

```python
df = df.drop(columns=['Sex'],axis='columns')
```

```python
df['Fare'] = df['Fare'].astype(int)
df.tail()
```

```python
x = df.drop(columns=['PassengerId','Survived','Pclass','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[33,5,1]])
```

```python
x1 = model.predict(x_test)
```

```python
import seaborn as sn
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, x1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel("Precision")
plt.ylabel("Truth")
```
