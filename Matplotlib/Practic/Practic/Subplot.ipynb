```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,1,2,3])
y = np.array([10,43,21,43])
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Income")
plt.show()
```
